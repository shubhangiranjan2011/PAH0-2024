```python
from mtranslate import translate

def translate_text(text):
    if isinstance(text, str):
        try:
            # Encode text to bytes before translation
            text_bytes = text.encode('utf-8')
            translated = translate(text_bytes, 'en', 'auto')
            return translated
        except Exception as e:
            print(f"Translation failed: {e}")
            return "Translation failed"
    else:
        return ""


translated_texts = []
for text in rawdrugnames:
    translated_texts.append(translate_text(text))
print(translated_texts)
```

```python
translated_texts[::10]
```

```python
drug_names = []
concentrations = []


for row in translated_texts[3:]:
  # Find the first index of a numeric character
  first_numeric_index = next((i for i, char in enumerate(row) if char.isdigit()), None)

  # Check for "OF" pattern (special case)
  of_index = row.lower().find('of')  # Convert to lowercase for case-insensitive search

  # Split based on "OF" or numeric character
  if of_index != -1:
    drug_name = row[of_index+3:].strip()  # Extract after "OF" (skip "OF ")
    composition = row[:of_index].strip()  # Extract before "OF"
    composition = composition.replace("IT'S", '')
  else:
    # Default logic for numeric separation
    if first_numeric_index is not None:
      drug_name = row[:first_numeric_index]
      composition = row[first_numeric_index:]
    else:
      drug_name = row
      composition = ""

  # Append the extracted values to the lists
  drug_names.append(drug_name)
  concentrations.append(composition)


# Create DataFrame
df = pd.DataFrame({ 'Drug Original Name': drug_names,
    'Concentration': concentrations
})

print(df)
```

```python
df.to_csv('PERantineoplastics2-cleaned.csv')
```
