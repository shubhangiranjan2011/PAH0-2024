```python
from mtranslate import translate

def translate_text(text):
    if isinstance(text, str):
        try:
            # Encode text to bytes before translation
            text_bytes = text.encode('utf-8')
            translated = translate(text_bytes, 'en', 'auto')
            return translated
        except Exception as e:
            print(f"Translation failed: {e}")
            return "Translation failed"
    else:
        return ""


translated_texts = []
for text in rawdrugnames:
    translated_texts.append(translate_text(text))
print(translated_texts)
```

```python
drug_names = []
concentrations = []
pharmaceutical_forms = []
comp = []

for item in translated_texts:
    parts = item.split(", ")
    drug_names.append(parts[0])
    if len(parts) > 1:
        if 'COMPOSITION' in parts[1]:
          comp.append(parts[1].split(": ")[1])
          concentrations.append(parts[2].split(":")[1] if 'CONCENTRATION' or 'DOSAGE' in parts[2] else None)
          if len(parts) > 2:
            if 'PHARMACEUTICAL FORM' in parts[2] or 'INDICATION' in parts[2] or 'PRESENTATION' in parts[2]:
              pharmaceutical_forms.append(parts[2].split(": ")[1])
            elif 'PHARMACEUTICAL FORM' in parts[3]:
              pharmaceutical_forms.append(parts[3].split(": ")[1])
          else:
            comp.append(None)
        else:
          comp.append(None)
          concentrations.append(parts[1].split(":")[1] if 'CONCENTRATION' or 'DOSAGE' in parts[1] else None)
          if len(parts) > 2:
            if 'PHARMACEUTICAL FORM' in parts[2] or 'INDICATION' in parts[2] or 'PRESENTATION' in parts[2]:
              pharmaceutical_forms.append(parts[2].split(": ")[1])
            elif 'PHARMACEUTICAL FORM' in parts[3]:
              pharmaceutical_forms.append(parts[3].split(": ")[1])
          else:
            pharmaceutical_forms.append(None)
    else:
      concentrations.append(None)
      pharmaceutical_forms.append(None)
      comp.append(None)


# Create DataFrame
df = pd.DataFrame({ 'Drug Original Name': drug_names,
    'Compositions': comp,
    'Concentration': concentrations,
    'Pharmaceutical Form': pharmaceutical_forms
})

print(df)
```
