```python
from mtranslate import translate

def translate_text(text):
    if isinstance(text, str):
        try:
            # Encode text to bytes before translation
            text_bytes = text.encode('utf-8')
            translated = translate(text_bytes, 'en', 'auto')
            return translated
        except Exception as e:
            print(f"Translation failed: {e}")
            return "Translation failed"
    else:
        return ""


translated_texts = []
for text in rawdrugnames:
    translated_texts.append(translate_text(text))
print(translated_texts)
```

```python
translated_texts
```

```python
drug_names = []
concentrations = []
pharmaceutical_forms = []
presentation = []

for item in translated_texts[2:]:
    parts = item.split(", ")
    drug_names.append(parts[0])
    if len(parts) > 1:
        if 'DOSAGE' in parts[1]:
          concentrations.append(parts[1].split(":")[1])

          if len(parts) > 2:
            if 'PRESENTATION' in parts[1] :
              presentation(parts[1].split(": ")[1])
            else:
              presentation.append(None)
            if 'DOSAGE' in parts[2]:
              concentrations.append(parts[2].split(": ")[1])
            if 'INDICATION' or 'PHARMACEUTICAL FORM' in parts[2]:
              pharmaceutical_forms.append(parts[2].split(":")[1])
            else:
              pharmaceutical_forms.append(None)

          else:
            presentation.append(None)
            pharmaceutical_forms.append(None)
        else:
          concentrations.append(None)
          presentation.append(None)
          pharmaceutical_forms.append(None)



# Create DataFrame
df = pd.DataFrame({ 'Drug Original Name': drug_names,'Concentration': concentrations, 'Presentation': presentation, 'Pharmaceutical Form': pharmaceutical_forms})

print(df)
```
