```python
rawdrugnames =  data[data.columns[1]]
rawdrugnames.tolist()
rawdrugnames
```

```python
from mtranslate import translate

def translate_text(text):
    if isinstance(text, str):
        try:
            # Encode text to bytes before translation
            text_bytes = text.encode('utf-8')
            translated = translate(text_bytes, 'en', 'auto')
            return translated
        except Exception as e:
            print(f"Translation failed: {e}")
            return "Translation failed"
    else:
        return ""


translated_texts = []
for text in rawdrugnames:
    translated_texts.append(translate_text(text))
print(translated_texts)
```

```python
translated_texts[1:50]
```

```python


# Extracting data
drug_names = []
compositions = []
concentrations = []
pharmaceutical_forms = []

for item in translated_texts[2:]:
    parts = item.split(', ')
    drug_name = parts[0]
    composition = None
    concentration = None
    pharmaceutical_form = None
    for part in parts[1:]:
        if part.startswith('ACTIVE INGREDIENT:') or part.startswith('COMPOSITION:'):
            composition = part.split(': ')[1]
        elif part.startswith('CONCENTRATION:') or part.startswith('DOSAGE:'):
            concentration = part.split(': ')[1]
        elif part.startswith('PHARMACEUTICAL FORM:') or part.startswith('PRESENTATION:'):
            pharmaceutical_form = part.split(': ')[1]
    drug_names.append(drug_name)
    compositions.append(composition)
    concentrations.append(concentration)
    pharmaceutical_forms.append(pharmaceutical_form)

# Creating DataFrame
df = pd.DataFrame({
    'Drug Name': drug_names,
    'Composition': compositions,
    'Concentration': concentrations,
    'Pharmaceutical Form': pharmaceutical_forms
})

# Replace empty strings with None
df.replace('', pd.NA, inplace=True)

print(df)
```

```python
df.to_csv('BRACardio3-cleaned.csv')
```
